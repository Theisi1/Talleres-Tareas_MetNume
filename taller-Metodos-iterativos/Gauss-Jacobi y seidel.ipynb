## Ejercicio 1


$x_1+x_2=7$ 

$−2x_1+5x_2=0$

* X_0 = (0, 0)

```python
A = np.array([
    [1,1],
    [-2,5]
], dtype=float)

B = np.array([
    [7],
    [0]
], dtype= float)

X = np.array([
    [0],
    [0]
],dtype= float)

n = len(B)

sol_Jacobi, aprox = gauss_jacobi(A = A, b = B, x = X * n, tol= 1e-5, max_iter=10 )
```

```python
sol_seidel, aprox_2 = gauss_seidel(A = A, b = B, x = X * n, tol= 1e-5, max_iter=10 )
```

```python
import matplotlib.pyplot as plt

x_s = np.linspace(-10, 10, 400)

def ec_1(x):  
    return 7 - x

def ec_2(x):  
    return (2/5) * x



xs = [v[0, 0] for v in aprox]
ys = [v[1, 0] for v in aprox]

plt.figure(figsize=(7, 7))

plt.plot(x_s, ec_1(x_s), label='x + y = 7')
plt.plot(x_s, ec_2(x_s), label='-2x + 5y = 0')

# Trayectoria de Jacobi 
plt.plot(xs, ys, marker='o', linestyle='-', linewidth=1, label='Aproximaciones Jacobi')

plt.scatter(sol_Jacobi[0, 0], sol_Jacobi[1, 0], s=80, marker='D')

plt.xlim(-5, 10)
plt.ylim(-5, 15)
plt.axhline(0, color='black', linewidth=0.8, ls='--')
plt.axvline(0, color='black', linewidth=0.8, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title("Sistema y aproximaciones por Jacobi")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

## Ejercicio 3

$$
A = \begin{bmatrix}
1 & -1 & 2 & 0 \\
-1 & 1 & -1 & 3 \\
2 & -1 & 1 & -1 \\
0 & 3 & -1 & 2
\end{bmatrix}
$$

$$
b = \begin{bmatrix}
1360 \\
1130 \\
1350 \\
3650
\end{bmatrix}
$$

- Para $(0,0,0,0)$

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([
    [ 1, -1,  2,  0],
    [-1,  1, -1,  3],
    [ 2, -1,  1, -1],
    [ 0,  3, -1,  2]
], dtype=float)

b = np.array([
    [1360],
    [1130],
    [1350],
    [3650]
], dtype=float)



x0 = np.zeros((4, 1), dtype=float)
n = len(B)


sol_Jacobi, aprox = gauss_jacobi(A=A, b=b, x=x0, tol=1e-5, max_iter=10)
```

```python
aprox_np = np.squeeze(np.array(aprox, dtype=float))  

k = np.arange(1, aprox_np.shape[0] + 1)

plt.figure(figsize=(8, 5))
plt.plot(k, aprox_np[:, 0], marker='o', linewidth=1, label=r"$x_1^{(k)}$")
plt.plot(k, aprox_np[:, 1], marker='o', linewidth=1, label=r"$x_2^{(k)}$")
plt.plot(k, aprox_np[:, 2], marker='o', linewidth=1, label=r"$x_3^{(k)}$")
plt.plot(k, aprox_np[:, 3], marker='o', linewidth=1, label=r"$x_4^{(k)}$")

plt.grid(True, linestyle='--', linewidth=0.5)
plt.xlabel("Iteración k")
plt.ylabel("Valor de la componente")
plt.title("Convergencia de Jacobi por componentes")
plt.legend()
plt.show()
```
