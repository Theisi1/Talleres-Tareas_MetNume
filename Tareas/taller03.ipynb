### Ejericio 1
- Grafique las curvas de las series de Taylor de varios órdenes para los soguientes casos:

1. 
$\ln(x)$,
$x_0 = 1$
$ n = 5$

$ \frac{1}{5}(x-1)^5 - \frac{1}{4}(x-1)^4 + \frac{1}{3}(x-1)^3 $
$ - \frac{1}{2}(x-1)^2 + (x-1) $




```python
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.linspace(0.1, 1.7, 400)

f = np.log(x)

x1 = (x - 1)
x2 = (x - 1) - (x - 1)**2 / 2
x3 = (x - 1) - (x - 1)**2 / 2 + (x - 1)**3 / 3
x4 = (x - 1) - (x - 1)**2 / 2 + (x - 1)**3 / 3 - (x - 1)**4 / 4
x5 = (x - 1) - (x - 1)**2 / 2 + (x - 1)**3 / 3 - (x - 1)**4 / 4 + (x - 1)**5 / 5


plt.figure(figsize=(8,6))
plt.plot(x, f, label='y = ln(x)', color='black', linewidth=2)
plt.plot(x, x1, '--', color='red', label='1er orden')
plt.plot(x, x2, '--', color='orange', label='2º orden')
plt.plot(x, x3, '--', color='green', label='3er orden')
plt.plot(x, x4, '--', color='blue', label='4º orden')
plt.plot(x, x5, '--', color='purple', label='5º orden')

plt.scatter(1, 0, color='black', s=60, zorder=5, label='x₀ = 1')
plt.text(1.02, 0.05, '(1, 0)', fontsize=10)

plt.title('Serie de Taylor de ln(x) en x₀ = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.7)
plt.axvline(1, color='gray', linestyle=':')
plt.show()

    
```

$ \frac{x^4}{24} - \frac{x^2}{2} $

```python
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.linspace(-6, 6, 400)


f = np.cos(x)


x0 = 0
cos0 = math.cos(x0)  

x1 = cos0 * np.ones_like(x)                                  
x2 = cos0 - (x - x0)**2 / math.factorial(2)                 
x3 = cos0 - (x - x0)**2 / math.factorial(2) + (x - x0)**4 / math.factorial(4)  
x4 = x3  
x5 = x3  


plt.figure(figsize=(8,6))
plt.plot(x, f, label='y = cos(x)', color='black', linewidth=2)
plt.plot(x, x1, '--', color='red', label='1er orden')
plt.plot(x, x2, '--', color='orange', label='2º orden')
plt.plot(x, x3, '--', color='green', label='3er orden / 5º grado')

plt.scatter(x0, cos0, color='black', s=60, zorder=5, label='x₀ = 0')
plt.text(x0 + 0.2, cos0 - 0.05, f'({x0}, {cos0:.1f})', fontsize=10)

plt.title('Serie de Taylor de cos(x) en x₀ = 0 (hasta grado 5)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axvline(0, color='gray', linestyle=':')
plt.show()



```

3. 

$ \frac{1}{1-x}, x_0 = 0.5, n = 5 $

$ P_5(x) = 2 + 4(x-\frac{1}{2}) + 8(x-\frac{1}{2})^2 + 16(x-\frac{1}{2})^3 + 32(x-\frac{1}{2})^4 + 64(x-\frac{1}{2})^5 $



```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-1, 1.5, 400)


f = 1 / (1 - x)

x0 = 0.5
f0 = 1 / (1 - x0)  

x1 = 2 * np.ones_like(x)  
x2 = 2 + 4*(x - x0)       
x3 = 2 + 4*(x - x0) + 8*(x - x0)**2      
x4 = 2 + 4*(x - x0) + 8*(x - x0)**2 + 16*(x - x0)**3  
x5 = 2 + 4*(x - x0) + 8*(x - x0)**2 + 16*(x - x0)**3 + 32*(x - x0)**4 + 64*(x - x0)**5  


plt.figure(figsize=(8,6))
plt.plot(x, f, label='y = 1/(1 - x)', color='black', linewidth=2)
plt.plot(x, x1, '--', color='red', label='1er orden')
plt.plot(x, x2, '--', color='orange', label='2º orden')
plt.plot(x, x3, '--', color='green', label='3er orden')
plt.plot(x, x4, '--', color='blue', label='4º orden')
plt.plot(x, x5, '--', color='purple', label='5º orden')

plt.scatter(x0, f0, color='black', s=60, zorder=5, label='x₀ = 1/2')
plt.text(x0 + 0.05, f0 + 0.2, f'({x0}, {f0:.1f})', fontsize=10)

plt.title('Serie de Taylor de f(x) = 1/(1 - x) en x₀ = 1/2 (grado 5)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axvline(x0, color='gray', linestyle=':')
plt.ylim(-5, 10)
plt.show()
```

### Ejercicio 2

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:

1. 
$ (0, 0), (30, \frac{1}{2}), (60, \frac{\sqrt{3}}{2}), (90, 1) $
$$
x_0 = 0,\quad x_1 = 30,\quad x_2 = 60,\quad x_3 = 90
$$

$$
y_0 = 0,\quad y_1 = 0.5,\quad y_2 = \tfrac{\sqrt{3}}{2},\quad y_3 = 1
$$

---

## Bases de Lagrange

###  $L_0(x)$ 

$$
L_0(x)=\frac{(x-30)(x-60)(x-90)}
{(0-30)(0-60)(0-90)}
=\frac{(x-30)(x-60)(x-90)}{162000}
$$

###  $L_1(x)$ 

$$
L_1(x)=\frac{x(x-60)(x-90)}
{(30-0)(30-60)(30-90)}
=\frac{x(x-60)(x-90)}{54000}
$$

###  $L_2(x)$ 

$$
L_2(x)=\frac{x(x-30)(x-90)}
{(60-0)(60-30)(60-90)}
=-\frac{x(x-30)(x-90)}{54000}
$$

###  $L_3(x)$ 

$$
L_3(x)=\frac{x(x-30)(x-60)}
{(90-0)(90-30)(90-60)}
=\frac{x(x-30)(x-60)}{162000}
$$

---

## Polinomio 

$$
P(x)=
0\,L_0(x)
+\frac{1}{2}L_1(x)
+\frac{\sqrt{3}}{2}L_2(x)
+1\cdot L_3(x)
$$



$$
P(x)=
\frac{1}{2}\left(\frac{x(x-60)(x-90)}{54000}\right)
+
\frac{\sqrt{3}}{2}\left(\frac{x(x-30)(x-90)}{54000}\right)
+
\left(\frac{x(x-30)(x-60)}{162000}\right)
$$



$$
P(x)=
-6.054\times 10^{-7}x^{3}
-1.9943\times 10^{-5}x^{2}
+0.0178098x
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math


xp = np.array([0, 30, 60, 90])
yp = np.array([0, 0.5, math.sqrt(3)/2, 1])


x = np.linspace(0, 90, 400)


L0 = (x-30)*(x-60)*(x-90) / 162000
L1 =  x*(x-60)*(x-90)     / 54000
L2 =  x*(x-30)*(x-90)     / 54000
L3 =  x*(x-30)*(x-60)     / 162000


P = 0*L0 + 0.5*L1 + (math.sqrt(3)/2)*L2 + 1*L3


plt.figure(figsize=(8,6))
plt.plot(x, P, label='Polinomio de Lagrange', color='purple', linewidth=2)
plt.scatter(xp, yp, color='black', s=60, zorder=5, label='Datos')


plt.text(0,   0,    "(0,0)")
plt.text(30,  0.5,  "(30,0.5)")
plt.text(60,  math.sqrt(3)/2, "(60,√3/2)")
plt.text(90,  1,    "(90,1)")

plt.title('Polinomio de Lagrange para los puntos dados')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Bases de Lagrange

### $L_0(x)$

$$
L_0(x)=\frac{(x-2)(x-3)}{(1-2)(1-3)}
      =\frac{(x-2)(x-3)}{2}
$$

### $L_1(x)$

$$
L_1(x)=\frac{(x-1)(x-3)}{(2-1)(2-3)}
      =-\,(x-1)(x-3)
$$

### $L_2(x)$

$$
L_2(x)=\frac{(x-1)(x-2)}{(3-1)(3-2)}
      =\frac{(x-1)(x-2)}{2}
$$

## Polinomio 

$$
P(x)=
1\cdot L_0(x)
+2\cdot L_1(x)
+2\cdot L_2(x)
$$


$$

P(x)= -\frac{1}{2}x^2 + \frac{5}{2}x - 1

$$



```python
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.linspace(0, 10, 200)


P = -0.5*x**2 + 2.5*x - 1


xp = np.array([1, 2, 3])
yp = np.array([1, 2, 2])

plt.figure(figsize=(8,6))
plt.plot(x, P, label='Polinomio de Lagrange', color='#FF00FF', linewidth=2)
plt.scatter(xp, yp, color='red', label='Puntos dados', zorder=3)
plt.axvline(0, color='black', linewidth=0.9)
plt.axhline(0, color='black', linewidth=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 7)
plt.ylim(0, 3)
plt.grid(True)
plt.legend()
plt.title('Interpolación de Lagrange')
plt.show()
```
