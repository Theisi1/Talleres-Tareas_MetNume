3. Encuentre el intervalo más largo en el que se debe encontrar $p^*$ para aproximarse a $p$ con error relativo máximo de $10^{-4}$ para cada valor $p$



```python
import math

def intervalo_aproximacion(p, error_rel_max=1e-4):
    if p == 0:
        return (-error_rel_max, error_rel_max)
    else:
        delta = abs(p) * error_rel_max
        return (p - delta, p + delta)
    

#  1: p = π
p = math.pi
intervalo_pi = intervalo_aproximacion(p)
print(f"Para p = π ≈ {p}")
print(f"Intervalo: [{intervalo_pi[0]:.10f}, {intervalo_pi[1]:.10f}]")

#Ejemplo 2: p = e
p = math.e
intervalo_e = intervalo_aproximacion(p)
print(f"\nPara p = e ≈ {p}")
print(f"Intervalo: [{intervalo_e[0]:.10f}, {intervalo_e[1]:.10f}]")

# 3: p = sqrt{2}

p = math.sqrt(2)
intervalo_sqr = intervalo_aproximacion(p)
print(f"\nPara p = sqrt(2) ≈ {p} ")
print(f"Intervalo: [{intervalo_sqr[0]:.10f}, {intervalo_sqr[1]:.10f}]")

# 4: p = raiz cubica de (7)
p = 7 ** (1/3)
intervalo_raiz7 = intervalo_aproximacion(p)
print(f"\nPara p = raíz cubica de 7 ≈ {p}")
print(f"Intervalo: [{intervalo_raiz7[0]:.10f}, {intervalo_raiz7[1]:.10f}]")

    
```

#### d. $\frac{\sqrt{13} + \sqrt{11}}{\sqrt{13} - \sqrt{11}} $

* $ \sqrt{13} = 3.60555 \approx 3.61 $
* $ \sqrt{11} = 3.31662 \approx 3.32 $
* Numerador $= 3.61 + 3.32 = 6.93 \approx 6.93$
* Denominador $= 3.61 - 3.32 = 0.29 \approx 0.29$
* Cociente $= \frac{6.93}{0.29} = 23.8966 \approx 23.9$

 Valor Exacto: p = 23.9582607431

Errores:
* $ Error_{abs} =  5.8261\times10^{-2}$
* $ Error_{rela} \approx 2.4318\times 10^{-3}$


```python
import math

def round_sig(x, sig=3):
    if x == 0: 
        return 0.0
    return round(x, sig - 1 - int(math.floor(math.log10(abs(x)))))

def b_three_digit():
    t1 = round_sig(-10*math.pi)
    t2 = round_sig(6*math.e)
    s = round_sig(t1 + t2)
    r = round_sig(3/61)
    val = round_sig(s - r)
    return val

def errors(approx, exact):
    ea = abs(approx - exact)
    er = ea/abs(exact) if exact != 0 else float("inf")
    return ea, er


B_ap, B_ex = b_three_digit(), -10*math.pi + 6*math.e - 3/61

for name, ap, ex in [('b',B_ap,B_ex)]:
    ea, er = errors(ap, ex)
    print(f"{name}) aprox={ap}, exacto={ex:.12f},  E_abs={ea:.12g},  E_rel={er:.12g}")
```

$ x = \frac{x_0y_1-x_1y_0}{y_1-y_0} $

Primer método:

$x = \frac{x_0y_1-x_1y_0}{y_1-y_0} $

Numerador: $7.55 - 6.25 = 1.30$

$x = x_0 - \frac{(x_1-x_0)y_0}{y_1-y_0} $

$I.$
