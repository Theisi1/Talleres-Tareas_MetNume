1. Use el m√©todo de bisecci√≥n para encontrar soluciones precisas de $10^{-2}$ para $x^3 - 7x^2 + 14x - 6 = 0$ en cada intervalo.

a. $[0,1]$

b. $[1, 3.2]$

c. $[3.2, 4]$


```python
from scipy.optimize import bisect

def f(x):
    return x**3 - 7*x**2 + 14*x - 6

intervalos = [(0, 1), (1, 3.2), (3.2, 4)]
tolerancia = 1e-2

for a, b in intervalos:
    raiz = bisect(f, a, b, xtol=tolerancia)
    print(f"Ra√≠z en el intervalo [{a}, {b}]: {raiz:.5f}")
```

5. Sea $f(x) = (x+3)(x+1)^2x(x-1)^3(x-3).$ ¬ø En qu√© cero de $f$ converge el m√©todo de bisecci√≥n cuando se aplica en los siguientes intervalos?

a. $[-1.5,2.5]$

b. $[-0.5,2.4]$

c. $[-0.5,3]$

d. $[-3,-0.5]$

```python
import numpy as np

def biseccion(f, a, b, TOL, N=1000):
    FA = f(a)
    FB = f(b)

    if abs(FA) < TOL:
        return a
    if abs(FB) < TOL:
        return b
    if FA * FB > 0:
        print(f"No hay cambio de signo en el intervalo [{a}, {b}]")
        return None

    for _ in range(N):
        p = (a + b) / 2
        FP = f(p)
        if abs(b - a) / 2 < TOL or abs(FP) < TOL:
            return p
        if FA * FP > 0:
            a, FA = p, FP
        else:
            b = p
    print(f"No se alcanz√≥ la tolerancia en {N} iteraciones.")
    return None


def f6(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)


intervalos = [
    (-1.5, 2.5),
    (-0.5, 2.4),
    (-0.5, 3),
    (-3, -0.5)
]

for idx, (a, b) in enumerate(intervalos, start=1):
    raiz = biseccion(f6, a, b, 1e-3)
    print(f"Intervalo {idx}) [{a}, {b}] ‚Üí Ra√≠z aproximada: {raiz}")
```
