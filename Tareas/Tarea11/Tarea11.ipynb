=

\begin{bmatrix}

110 \\ 110 \\ 220 \\ 110 \\ 110 \\ 220
\end{bmatrix}

a. ¬øLa matriz es estrictamente diagonalmente dominante?

S√≠. La matriz es estrictamente diagonalmente dominante por filas.


$|a_{ii}| > \sum_{j \ne i} |a_{ij}|$


En este caso, para todas las filas se tiene:


$|a_{ii}| = 4$


y la suma de los valores absolutos de los elementos fuera de la diagonal es:


$\sum_{j \ne i} |a_{ij}| = 1 + 1 = 2$


Como se cumple que


$4 > 2$


en todas las filas, se concluye que la matriz es estrictamente diagonalmente dominante.
