1. Utilice aritmética de corte de tres dígitos para calcular las siguientes sumas. Para cada parte, ¿qué método es más preciso y por qué?

a. $ \sum_{i = 1}^{10} (\frac{1}{i^2})$ primero por $ \frac{1}{1} + \frac{1}{4} + ... + \frac{1}{100} $ y luego por $ \frac{1}{100} + \frac{1}{81} + ... + \frac{1}{1}  $

- Ascendente 

$ \sum_{i = 1}^{10} (\frac{1}{i^2}) = \frac{1}{1^2} +\frac{1}{2^2} + \frac{1}{3^2} + \frac{1}{4^2} + \frac{1}{5^2} + \frac{1}{6^2} +  \frac{1}{7^2} + \frac{1}{8^2} + \frac{1}{9^2} + \frac{1}{10^2}$

$ = 1.53 $

-Descendente

$ \sum_{i = 1}^{10} (\frac{1}{i^2}) = \frac{1}{10^2} +\frac{1}{9^2} + \frac{1}{8^2} + \frac{1}{7^2} + \frac{1}{6^2} + \frac{1}{5^2} +  \frac{1}{4^2} + \frac{1}{3^2} + \frac{1}{2^2} + \frac{1}{1^2}$

$ = 1.54 $

Análisis: 

El método descendente es más preciso porque al sumar primero los valores grandes y luego los pequeños, se reduce la pérdida de cifras significativas y el error acumulado por trucamiento. En cambio, el método ascendente acumula más error al sumar términos pequeños a resultados grandes.


```python
import math

def corte_tres_digitos(x):
    if x == 0:
        return 0.0
    signo = 1 if x > 0 else -1
    x = abs(x)
    exp = math.floor(math.log10(x))
    k = 10 ** (3 - 1 - exp)
    return signo * math.floor(x * k) / k

# Ascendente
suma_asc = 0
for n in range(1, 11):
    termino = 1 / (n ** 2)
    termino = corte_tres_digitos(termino)
    suma_asc = corte_tres_digitos(suma_asc + termino)

# Descendente
suma_desc = 0
for n in range(10, 0, -1):
    termino = 1 / (n ** 2)
    termino = corte_tres_digitos(termino)
    suma_desc = corte_tres_digitos(suma_desc + termino)

print("Resultado Orden Ascendente con corte a tres dígitos:", suma_asc)
print("Resultado Orden Descendente con corte a tres dígitos:", suma_desc)

```

b. $ \sum_{i = 1}^{10} (\frac{1}{i^2})$ primero por $ \frac{1}{1} + \frac{1}{8} + \frac{1}{27} + ... + \frac{1}{1000} $ y luego por $ \frac{1}{1000} + \frac{1}{729} + ... + \frac{1}{1}  $

- Ascendente 

$ \sum_{i = 1}^{10} (\frac{1}{i^3}) = \frac{1}{1^3} +\frac{1}{2^3} + \frac{1}{3^3} + \frac{1}{4^3} + \frac{1}{5^3} + \frac{1}{6^3} +  \frac{1}{7^3} + \frac{1}{8^3} + \frac{1}{9^3} + \frac{1}{10^3}$

$ = 1.16 $

-Descendente

$ \sum_{i = 1}^{10} (\frac{1}{i^3}) = \frac{1}{10^3} +\frac{1}{9^3} + \frac{1}{8^3} + \frac{1}{7^3} + \frac{1}{6^3} + \frac{1}{5^3} +  \frac{1}{4^3} + \frac{1}{3^3} + \frac{1}{2^3} + \frac{1}{1^3}$

$ = 1.19 $

Análisis:

El método descendente es más preciso porque al sumar primero los valores grandes y después los pequeños se minimiza la pérdida de cifras significativas y el error de truncamiento.
En cambio, en el método ascendente los términos pequeños se pierden al sumarse con una cantidad mayor, provocando un resultado menos exacto.


```python
import math

def corte_tres_digitos(x):
    if x == 0:
        return 0.0
    signo = 1 if x > 0 else -1
    x = abs(x)
    exp = math.floor(math.log10(x))
    k = 10 ** (3 - 1 - exp)
    return signo * math.floor(x * k) / k

# Ascendente
suma_asc = 0
for n in range(1, 11):
    termino = 1 / (n ** 3)
    termino = corte_tres_digitos(termino)
    suma_asc = corte_tres_digitos(suma_asc + termino)

# Descendente
suma_desc = 0
for n in range(10, 0, -1):
    termino = 1 / (n ** 3)
    termino = corte_tres_digitos(termino)
    suma_desc = corte_tres_digitos(suma_desc + termino)

print("Resultado Orden Ascendente con corte a tres dígitos:", suma_asc)
print("Resultado Orden Descendente con corte a tres dígitos:", suma_desc)
```

2. La serie de Maclaurin para la función arcotangente converge para $ -1 < x \leq 1 $ y esta dada por 

$ \arctan x = \lim_{n \to \infty} P_n(x) = \lim_{n \to \infty} \sum_{i = 1}^{n} (-1)^{i+1} \frac{x^{2i-1}}{2i - 1}  $ 

a. Utilice el hecho de que tan $\frac{\pi}{4} = 1 $ para determinar el número n de términos de la serie que se necesita sumar para garantizar que $|4P_n(1)-\pi| < 10^{-3} $

```python
import math

def arct_macl(n):
    acumular = 0
    for k in range(1, n + 1 ):
        termino = ((-1)**(k+1)) / (2 * k - 1)
        acumular += termino
    return acumular

eps = 1e-3
err= 1
n = 0

while err >= eps:
    n += 1
    aprox_pi = 4 * arct_macl(n)
    err = abs(aprox_pi - math.pi)
    
print("Número minimo de terminos necesarios:" , n)
```

b. El lenguaje de programación C++ requiere que el valor de $\pi$ se encuentre dentro de $10^{-10}$. ¿ Cuantos términos de la serie se necesitarían sumar para obtener este grado de prescisión?

```python
import math

def arctan_maclaurin(n):
    suma = 0
    for k in range(1, n + 1):
        suma += ((-1) ** (k + 1)) / (2 * k - 1)
    return suma


epsilon = 1e-10 # Va a tardar mucho por el orden de magnitud.
error = 1
n = 0

while error >= epsilon:
    n += 1
    aprox_pi = 4 * arctan_maclaurin(n)
    error = abs(aprox_pi - math.pi)

print("Número mínimo de términos necesarios:", n)
```

3. Otra fórmula para calcular $\pi$ se puede deducir a partir de la identidad $\frac{\pi}{4} = 4\arctan{\frac{1}{5}} - \arctan{\frac{1}{239}} $.
Determine el número de términos que se deben sumar para garantizar una aproximación $\pi$ dentro de $10^{-3} $

```python
import math

def aproximar_arctan(x, n):
    return sum(((-1) ** k) * (x ** (2 * k + 1)) / (2 * k + 1) for k in range(n))

def aproximar_pi_por_machin(n):
    parte1 = aproximar_arctan(1 / 5, n)
    parte2 = aproximar_arctan(1 / 239, n)
    return 4 * (4 * parte1 - parte2)

tolerancia = 1e-3
error = 1
terminos = 0

while error >= tolerancia:
    terminos += 1
    pi_aprox = aproximar_pi_por_machin(terminos)
    error = abs(pi_aprox - math.pi)

print(f"Cantidad mínima de términos requeridos para una precisión de 10^-3: {terminos}")

```

2. Las ecuaciones $(1.2)$ y $(1.3)$ en la sección $1.2$ proporcionan formas alternativas para las raíces $x_1$ y $x_2$ de $ax^2+bx+c = 0$
Construya un algoritmo con entrada $a, b, c$ y salida $x_1, x_2$ que calcule las raíces $x_1$ y $x_2$ (que pueden ser iguales o conjugadas complejas) mediante la mejor fórmula para cada raíz.

```python
import cmath

def resolver_raices(a, b, c):
    
    delta = b**2 - 4 * a * c

    # Caso 1: raíces reales distintas
    if delta > 0:
        if b > 0:
            r1 = (-b - delta**0.5) / (2 * a)
        else:
            r1 = (-b + delta**0.5) / (2 * a)
        r2 = c / (a * r1)

    # Caso 2: raíces reales e iguales
    elif delta == 0:
        r1 = r2 = -b / (2 * a)

    # Caso 3: raíces complejas conjugadas
    else:
        raiz_complex = cmath.sqrt(delta)
        r1 = (-b + raiz_complex) / (2 * a)
        r2 = (-b - raiz_complex) / (2 * a)

    return r1, r2



coef_a, coef_b, coef_c = 1, -4, 4
x1, x2 = resolver_raices(coef_a, coef_b, coef_c)

print(f"Las soluciones de la ecuación son: x1 = {x1}, x2 = {x2}")
```

3. Suponga que 

$\frac{1-2x}{1-x+x^2} + \frac{2x-4x^3}{1-x^2+x^4} + \frac{4x^3-8x^7}{1-x^4+x^8}+ ... = \frac{1+2x}{1+x+x^2}$,

para $x<1$ y si $x = 0.25$. Escriba y ejecute un algoritmo que determine el número de términos necesarios en el lado izquierdo de la ecuación de tal forma que el lado izquierdo difiera del lado derecho en menos de $10^{-6}$


```python
import math

def lado_derecho(x):
    return (1 + 2 * x) / (1 + x + x**2)

def aproximar_serie(x, tolerancia):
    suma = 0.0
    n = 1
    coef = 1
    potencia_y = x
    inv_x = 1 / x
    valor_real = lado_derecho(x)

    while True:
        numerador = coef * (potencia_y * inv_x) * (1 - 2 * potencia_y)
        denominador = 1 - potencia_y + potencia_y**2
        termino = numerador / denominador

        suma += termino

        # Verifi del error con respecto al lado derecho
        diferencia = abs(suma - valor_real)
        if diferencia < tolerancia:
            break

    
        coef *= 2
        potencia_y **= 2
        n += 1

    return n, suma, valor_real

# Datos del ejercicio
x = 0.25
tolerancia = 1e-6

n_terminos, suma_izquierda, lado_derecho_valor = aproximar_serie(x, tolerancia)

# Resultados
print(f"Número mínimo de términos requeridos: {n_terminos}")
print(f"Suma del lado izquierdo: {suma_izquierda}")
print(f"Valor del lado derecho: {lado_derecho_valor}")
print(f"Diferencia absoluta: {abs(suma_izquierda - lado_derecho_valor)}")
```
